# 泰坦尼克乘客生存预测
通过网站Kaggle上，提供819名乘客信息及是否生还的数据集，预测第二个数据集中乘客的生还情况。

## 背景

泰坦尼克号（RMS Titanic），又译作铁达尼号，是英国白星航运公司下辖的一艘奥林匹克级邮轮，于1909年3月31日在爱尔兰贝尔法斯特港的哈兰德与沃尔夫造船厂动工建造，1911年5月31日下水，1912年4月2日完工试航。
泰坦尼克号是当时世界上体积最庞大、内部设施最豪华的客运轮船，有“永不沉没”的美誉 。然而讽刺的是，在她的处女航中，泰坦尼克号便遭厄运——她从英国南安普敦出发，途经法国瑟堡-奥克特维尔以及爱尔兰昆士敦，驶向美国纽约，船上时间1912年4月14日23时40分左右，泰坦尼克号与一座冰山相撞，造成右舷船艏至船中部破裂，五座水密舱进水。次日凌晨2时20分左右，泰坦尼克船体断裂成两截后沉入大西洋底3700米处。2224名船员及乘客中，逾1500人丧生，其中仅333具罹难者遗体被寻回。泰坦尼克号沉没事故为和平时期死伤人数最惨重的海难之一，其残骸直至1985年才被再度发现，目前受到联合国教育、科学及文化组织的保护。
著名的数据分析竞赛网站Kaggle上，提供了一个“泰坦尼克号乘客生还预测”的题目。它有两个数据集，一个数据集中有819名乘客的信息以及他们最终是否生还，另一个数据集中记录了另外418名乘客信息，需要去预测第二个数据集中乘客的生还情况。

数据分析竞赛网站Kaggle上，提供数据集中包含的信息如下：
PassengerId          乘客编号
Survived             存活情况（存活：1 ; 死亡：0）
Pclass               客舱等级
Name                 乘客姓名
Sex                  性别
Age                  年龄
SibSp                同乘的兄弟姐妹/配偶数
Parch                同乘的父母/小孩数
Ticket               船票编号
Fare                 船票价格
Cabin                客舱号
Embarked             登船港口
已知一个带有标签（最终生还结果）的数据集，通过对该数据集进行分析建模，用得到的模型去对另一个数据集进行预测，预测的结果为两个：1-生还、0-死亡。

### 数据类型转换

由于要使用K最近邻算法，所有输入特征必须为数值型，需要将参与建模的特征转换为数值型

```python
train_dataset['Sex_1'] = train_dataset['Sex'].apply(lambda x: 1 if x == 'female' else 2 if x == 'male' else 0)
```

```python
def cabin_converter(v):
    m = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': '6', 'G': 7, 'T': 8}
    if v[0] in m:
        return int(m[v[0]])
    else:
        return 0

train_dataset['Cabin_2'] = train_dataset['Cabin_1'].apply(cabin_converter)
```

```python
def embarke_converter(v):
    m = {'C': 1, 'Q': 2, 'S': 3}
    if v in m:
        return m[v[0]]
    else:
        return 0

train_dataset['Embarked_1'] = train_dataset['Embarked'].apply(embarke_converter)
```

## 特征工程

### 特征生成

仔细观察数据，发现一些数据本身的特点，基于这些特点生成一些衍生的特征，或者仔细理解业务场景，基于业务场景的特点结合数据，产生一些衍生的特征。
通过观察数据，看到name（名字）字段中，有类似于Miss、Master、Mr等体现某种社会地位的信息，是否社会地位和能否逃生相关？可以考虑提取该特征title作为扩展特征。
通过观察数据，看到name（名字）字段中，有类似于Miss、Master、Mr等体现某种社会地位的信息，是否社会地位和能否逃生相关？可以考虑提取该特征title作为扩展特征。
观察到数据中Parch、Sibsp，可以计算家庭规模，家庭规模比较大的，由于互助的原因，有可能逃生几率更大，可以考虑衍生家庭规模size作为扩展特征。
另外还有一种情况，同一张船票多人使用，意味着多人结伴上船，逃生几率会大？可以考虑增加特征值同行人数partner作为衍生特征。

```python
import re
```

```python
def name_converter(v):
    m = {'Col': 1, 'Don': 2, 'Dr': 3, 'Jonkheer': 4, 'Lady': 5, 'Major': 6,
         'Master': 7, 'Miss': 8, 'Mlle': 9, 'Mme': 10, 'Mr': 11, 'Mrs': 12,
         'Ms': 13, 'Rev': 14, 'Sir': 15, 'the Countess': 16}
    for k in m:
        if re.search(k + '.', v):
            return m[k]
    return 0

train_dataset['Title'] = train_dataset['Name'].apply(name_converter)
```

```python
train_dataset['Size'] = train_dataset['SibSp'] + train_dataset['Parch']
```

```python
partners = train_dataset.groupby('Ticket').size()
train_dataset['Partner'] = train_dataset['Ticket'].map(lambda x: partners[x])
```

```python
train_dataset.head(10)
```

### 归一化处理

常见归一化算法有：min-max标准化和Z-score标准化方法。
min-max标准化也称为离差标准化，是对原始数据的线性变换，使结果值映射到[0 - 1]之间。
Z-score标准化将原始数据处理为符合标准正态分布，即均值为0，标准差为1。这种方式要求原始数据的分布可以近似为正态分布，否则处理的效果会变差。

```python
train_dataset_1 = train_dataset[['Pclass', 'SibSp', 'Parch', 'Fare',
                                    'Age_1', 'Sex_1', 'Embarked_1', 'Cabin_2',
                                    'Title', 'Size', 'Partner']]
train_dataset_norm = (train_dataset_1 - train_dataset_1.min()) / (train_dataset_1.max() - train_dataset_1.min())
```

```python
train_dataset_norm.head(10)
```

### 特征降维

特征维数（个数）过多，会带来各种问题，最重要的是可能会导致过拟合：在训练样本上表现很好，但在面对新的待预测样本的时候表现糟糕。需要使用降维的手段，减少特征值维数。另外，如果不同特征之间存在较强的关联性，也可以通过PCA降维减小对最终结果的影响。

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components='mle')
train_dataset_pca = pca.fit_transform(train_dataset_norm)
```

```python
train_dataset_pca.shape
```

【注】事实上本例的特征相互之间的关联性不大，没必要降维，后续继续使用PCA处理之前的数据进行训练。

### 特征评估和选择

对特征进行评估，选出重要性排名更靠前的特征。
这里使用随机森林来对特征的重要性进行度量。
1. 特征重要性度量
计算某个特征的重要性，步骤如下：
1）对每一颗决策树，选择相应的袋外数据（out of bag，OOB）计算袋外数据误差，记为errOOB1.
所谓袋外数据是指，每次建立决策树时，通过重复抽样得到一个数据用于训练决策树，这时还有大约1/3的数据没有被利用，没有参与决策树的建立。这部分数据可以用于对决策树的性能进行评估，计算模型的预测错误率，称为袋外数据误差。这已经经过证明是无偏估计的,所以在随机森林算法中不需要再进行交叉验证或者单独的测试集来获取测试集误差的无偏估计。
2）随机对袋外数据OOB所有样本的特征X加入噪声干扰（可以随机改变样本在特征X处的值），再次计算袋外数据误差，记为errOOB2。
3）假设森林中有N棵树，则特征X的重要性=∑（errOOB2-errOOB1）/N。这个数值之所以能够说明特征的重要性是因为，如果加入随机噪声后，袋外数据准确率大幅度下降（即errOOB2上升），说明这个特征对于样本的预测结果有很大影响，进而说明重要程度比较高。
2. 特征选择
在特征重要性的基础上，特征选择的步骤如下：
1）计算每个特征的重要性，并按降序排序
2）确定要剔除的比例，依据特征重要性剔除相应比例的特征，得到一个新的特征集
3）用新的特征集重复上述过程，直到剩下m个特征（m为提前设定的值）。
4）根据上述过程中得到的各个特征集和特征集对应的袋外误差率，选择袋外误差率最低的特征集。​

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, oob_score=True)
rf.fit(train_dataset_1, train_dataset['Survived'])
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]   # 降序排列
importance_features = [train_dataset_norm.columns[i] for i in indices]
importances = [importances[i] for i in indices]
print(importance_features)
print(importances)
```

## 算法建模和评估

### 模型训练

1.将数据集分成训练数据集合测试数据集
2.将训练数据集进行归一化
3.使用训练数据集的均值和方差将测试数据集归一化
4.使用训练数集训练模型
5.使用归一化后的测试数据集测试分类的准确度（accuracy）
6.使用网格搜索寻找最好的超参数，然后回到1-5

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, f1_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    train_dataset_1, train_dataset['Survived'], test_size=0.2)
```

```python
standardScaler = StandardScaler()
standardScaler.fit(x_train)
x_train = standardScaler.transform(x_train)
x_test = standardScaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

```python
y_predict = knn.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
```

```python
f1_score(y_test, y_predict)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid =[
    {
        'weights':['uniform'],
        'n_neighbors': [i for i in range(1,60)]
    },
    {
        'weights':['distance'],
        'n_neighbors': [i for i in range(1,60)],
        'p': [i for i in range(1,6)]
    }
]
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5, n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
knn_best = grid_search.best_estimator_
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
y_predict = knn_best.predict(x_test)
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
print(f1_score(y_test, y_predict))
```
